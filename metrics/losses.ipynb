```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mlp_path = 'mlp/'
neural_path = 'neuraltree/'

mlp_train_loss = list(np.load(mlp_path + 'mlp_train_losses.npy'))
mlp_val_loss = list(np.load(mlp_path + 'mlp_val_losses.npy'))
mlp_val_acc = list(np.load(mlp_path + 'mlp_val_accs.npy'))

neural_train_loss = list(np.load(neural_path + '30epoch_train_losses.npy'))
neural_val_loss = list(np.load(neural_path + '30epoch_val_losses.npy'))
neural_val_acc = list(np.load(neural_path + '30epoch_val_accs.npy'))
```

```python
mlp_train_loss = mlp_train_loss[30:]
mlp_val_loss = mlp_val_loss[30:]
mlp_val_acc = mlp_val_acc[30:]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(mlp_train_loss, label='MLP Training Loss')
plt.plot(mlp_val_loss, label='MLP Validation Loss')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Losses for MLP decoder')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(neural_train_loss, label='Neural Tree Training Loss')
plt.plot(neural_val_loss, label='Neural Tree Validation Loss')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Losses Neural Tree Decoder')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(mlp_val_acc, label='MLP Validation Accuracy')
plt.plot(neural_val_acc, label='Neural Tree Validation Accuracy')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Validation Accuracy')
plt.show()
```
